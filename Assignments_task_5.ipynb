```python
def media_post_analyzer(post_likes):
  total_likes = 0
  for likes in post_likes:
    total_likes += likes
  print(f"total likes: {total_likes}")
  if total_likes >=1000:
    print("viral post")
  else:
    print("not viral post")
media_post = [200,300,400,500]
media_post_analyzer(media_post)
```

```python
def check_rainfall(season_rain, minimum_required):
    sum_of_rain = 0
    for value in season_rain:
        sum_of_rain += value
    average = sum_of_rain / len(season_rain)
    print("Average Rainfall:", int(average))
    if average >= minimum_required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rain_data = [65, 78, 72, 69, 76]
required_level = 70
check_rainfall(rain_data, required_level)
```

```python
def check_usernames(usernames):
    seen_names = set()
    has_duplicate = False
    for name in usernames:
        if name in seen_names:
            has_duplicate = True
            break
        seen_names.add(name)
    if has_duplicate:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
user_data = ["userA", "userB", "userC", "userA"]
check_usernames(user_data)
```

```python
def verify_appointment(age_value):
    print("Patient Age:", age_value)
    if age_value >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")
age_input = 21
verify_appointment(age_input)
```

```python
def identify_premium_prices(price_values, limit_value):
    high_value_crops = []
    for price in price_values:
        if price > limit_value:
            high_value_crops.append(price)
    print("Premium Crops:", high_value_crops)
prices = [2100, 1800, 2500, 1999, 3200]
premium_threshold = 2000
identify_premium_prices(prices, premium_threshold)
```

```python
def check_application_status(error_count):
    print("Error Count:", error_count)
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("System Status:", status)
errors_detected = 7
check_application_status(errors_detected)
```

```python
def check_daily_limit(transaction_value):
    maximum_allowed = 50000
    print("Transaction Amount:", transaction_value)
    if transaction_value > maximum_allowed:
        print("Transaction Status: Rejected")
    else:
        print("Transaction Status: Approved")
amount_today = 60000
check_daily_limit(amount_today)
```

```python
def check_exam_eligibility(attendance_data):
    total_sessions = len(attendance_data)
    present_count = 0
    for entry in attendance_data:
        if entry == 1:
            present_count = present_count + 1
    attendance_percent = (present_count / total_sessions) * 100
    print("Attendance Percentage:", attendance_percent)
    if attendance_percent >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
# sample input(1 = Present, 0 = Absent)
record = [1, 1, 0, 1, 1]
check_exam_eligibility(record)
```

```python
def electricity_bill(units):
    total_bill = 0  # One step in (it's inside the function)

    if units <= 100:
        total_bill = units * 3
    elif units <= 200:
        total_bill = (100 * 3) + ((units - 100) * 5)
    else:
        total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    print(f"Units Consumed: {units}")
    print(f"Total Bill: ₹{total_bill}")

    if total_bill < 500:
        print("Status: Low Usage")
    elif 500 <= total_bill <= 1500:
        print("Status: Moderate Usage")
    else:
        print("Status: High Usage")

electricity_bill(250)
```
