```python
def smart_parking_lot(logs, capacity):
    current_count = 0
    peak_usage = 0

    for log in logs:

        if log == "IN":
            current_count += 1

            if current_count > peak_usage:
                peak_usage = current_count

        elif log == "OUT":
            if current_count > 0:
                current_count -= 1
            else:
                print("Invalid OUT detected")

    print("Currently Parked Vehicles:", current_count)
    print("Peak Usage:", peak_usage)

    if current_count > capacity:
        print("Parking Status: Exceeded")
    else:
        print("Parking Status: Available")

logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
capacity = 50

smart_parking_lot(logs, capacity)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 2

    # Traffic Adjustment
    traffic_delay = 0
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20

    # Weather Adjustment
    weather_delay = 0
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 15
    elif weather == "Storm":
        weather_delay = 25

    final_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", final_time, "minutes")

# Example usage:
estimate_delivery_time(10, "Medium", "Rainy")
estimate_delivery_time(5, "Low", "Clear")
estimate_delivery_time(20, "High", "Storm")
```

```python
def seat_occupancy_analyzer(total_seats, booked_seats):

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    # Determine status first
    if occupancy >= 100:
        status = "Housefull"
        suggestion = "Open Additional Show"

    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Consider Opening Additional Show"

    else:
        status = "Seats Available"
        suggestion = "No Need for Additional Show"

    # Print results at the end
    print(f"Occupancy: {round(occupancy)}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")


# Sample Input
total_seats = 200
booked_seats = [1] * 150

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):

    average = sum(cpu_readings) / len(cpu_readings)

    if average >= 80:
        status = "Critical"
    elif average >= 50:
        status = "Warning"
    else:
        status = "Normal"

    print(f"Average CPU Load: {round(average)}%")
    print(f"Server Status: {status}")


cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

```python
def classroom_resource_monitor(resource_usage):

    overused = [resource for resource, hours in resource_usage.items() if hours > 8]

    energy_alert = "Yes" if overused else "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", energy_alert)


resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

```python
def event_registration_controller(capacity, total_registrations):

    confirmed = min(capacity, total_registrations)
    waitlisted = max(0, total_registrations - capacity)

    status = "Closed" if confirmed >= capacity else "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
