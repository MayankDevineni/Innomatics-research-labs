Task 1: Conceptual Understanding

Q1. What is the difference between "Love" and "love" in NLP?

Answer:
In NLP, "Love" and "love" are treated as two distinct tokens because most tokenization processes are case-sensitive by default. This leads to redundant vocabulary and sparse feature representation, which can negatively impact model performance.

For example, a model may fail to recognize both as the same word in sentiment analysis. To avoid this, lowercasing is a standard preprocessing step.

Q2. What happens if stopwords are not removed?

Answer:
Stopwords such as "the", "is", "this", and "a" appear very frequently but carry little meaningful information.

If they are not removed:

They dominate vector representations like TF-IDF or Bag-of-Words
Important words get overshadowed
The model focuses on noise instead of signal

This can significantly reduce performance in tasks like text classification and sentiment analysis.

Q3. Provide two real-world scenarios where removing stopwords can be harmful.

Answer:

Sentiment Analysis with Negation
Example: "This is not bad" vs "This is bad"
Removing "not" completely changes the meaning
Leads to incorrect sentiment prediction
Voice Assistants / Search Queries
Example: "How do I turn this off?"
Removing stopwords → "turn off"
The intent becomes unclear

Short queries depend heavily on function words, so removing them can break understanding.

Q4. Explain the difference between stemming and lemmatization.

Answer:

Both techniques reduce words to their base form but differ in approach:

Stemming
Rule-based and fast
Removes suffixes blindly
Example: "caring" → "car" (incorrect)
Lemmatization
Uses vocabulary and grammar rules
Produces meaningful base words
Example: "caring" → "care", "worse" → "bad"

Task 2: Preprocessing

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (keep words)
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

Task 3: Stress Testing


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    results.append((sentence, tokens, clean))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Cleaned:", r[2])
```

Task 4: Token Analytics

```python
import numpy as np

for sentence, tokens, clean in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    print("\nSentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
```

Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
pipeline_output = full_pipeline(test_sentences)
pipeline_output
```

```python
print("TOKENS:")
for i, tokens in enumerate(pipeline_output["tokens"], 1):
    print(f"Sentence {i}: {tokens}")

print("\nCLEAN SENTENCES:")
for i, sentence in enumerate(pipeline_output["clean_sentences"], 1):
    print(f"Sentence {i}: {sentence}")
```

Task 7: Error Handling

```python
error_test_cases = [
    "",
    "😂😂😂😂😂",
    "1234567890",
    None
]

for case in error_test_cases:
    tokens, clean_sentence = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 60)
```
