```python
# 2. CONFIGURATION
import os
from google.colab import userdata
from langchain_groq import ChatGroq

# Securely load your key from Colab Secrets
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

# Initialize the LLM (Using Llama 3 via Groq for high performance)
llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0.5)
```

```python
# 3. PROMPTS & LCEL CHAINS
from langchain_core.prompts import ChatPromptTemplate

# Building a modular prompt
template = ChatPromptTemplate.from_messages([
    ("system", "You are a technical documentation expert."),
    ("user", "Summarize the benefits of LangChain for {audience}.")
])

# Modern LCEL syntax (Pipe operator)
chain = template | llm

print(chain.invoke({"audience": "Data Science Interns"}).content)
```

```python
# 4. CONVERSATIONAL MEMORY (LCEL)
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory

store = {}

def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

# Wrap chain with history
with_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="history",
)

# Execution to demonstrate state retention
config = {"configurable": {"session_id": "intern_session_1"}}
with_history.invoke({"input": "Hi, I am Mayank, working on a LangChain blog."}, config=config)
print(with_history.invoke({"input": "What project am I working on?"}, config=config).content)
```

```python
def validate_blog_input(topic, word_count):
    # Check for empty strings
    if not topic.strip():
        return False, "Error: Topic cannot be empty."

    # Ensure word count is a reasonable number
    try:
        count = int(word_count)
        if count < 100 or count > 2000:
            return False, "Error: Word count must be between 100 and 2000."
    except ValueError:
        return False, "Error: Word count must be a number."

    return True, "Success"

# User inputs
u_topic = input("Enter Blog Topic: ")
u_words = input("Enter Word Count: ")

is_valid, message = validate_blog_input(u_topic, u_words)

if is_valid:
    print(f"Input Validated. Proceeding to generate blog for: {u_topic}")
else:
    print(f" {message}")
```

```python
from langchain_core.prompts import PromptTemplate

# Define the 'Template' for the app
prompt_template = PromptTemplate.from_template(
    "You are a professional technical writer. Write a {length} word blog post "
    "about {topic} for a {audience} audience. "
    "Focus on: {focus_area}."
)

# Function to act as our 'Generator App'
def generate_prompt(topic, length, audience, focus):
    return prompt_template.format(
        topic=topic,
        length=length,
        audience=audience,
        focus_area=focus
    )

# Example usage
final_prompt = generate_prompt("LangChain Agents", "500", "Students", "ReAct Logic")
print("\n--- Generated Prompt for the LLM ---")
print(final_prompt)
```
