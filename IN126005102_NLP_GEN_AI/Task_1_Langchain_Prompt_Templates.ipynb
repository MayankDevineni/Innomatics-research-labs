Task 4: ChatPromptTemplate System

```python
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher."
        user_message = "Teach me about {topic}"

    elif role == "interviewer":
        system_message = "You are a technical interviewer."
        user_message = "Ask questions about {topic}"

    elif role == "motivator":
        system_message = "You are a motivational coach."
        user_message = "Motivate me to learn {topic}"

    else:
        system_message = "You are a helpful assistant."
        user_message = "Explain {topic}"

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", user_message)
    ])

    return chat_template.format_messages(topic=topic)


# Test
messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience. Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Defaulting to 'formal'")
        tone = "formal"

    return audience, tone


# Test
audience, tone = validate_inputs("kids", "cool")
print(audience, tone)
```

Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} clearly step by step for {audience} in a {tone} tone"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask 3 interview questions about {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

    else:
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone"
        )

    return template.format(topic=topic, audience=audience, tone=tone)


# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

Task 7: Template Reusability Test

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 Different Inputs
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("DSA", "expert", "casual"),
    ("Machine Learning", "beginner", "formal"),
    ("Cloud Computing", "intermediate", "fun")
]

for topic, audience, tone in inputs:
    print(reusable_template.format(topic=topic, audience=audience, tone=tone))
```
