```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF" # Default state

    def set_status(self, action):
        # We convert action to uppercase to handle 'on', 'ON', or 'On'
        if action.upper() == "ON":
            self.status = "ON"
        elif action.upper() == "OFF":
            self.status = "OFF"
        else:
            self.status = "INVALID ACTION"

    def display(self):
        print(f"{self.name} is {self.status}")

# --- Execution ---
name_input = "Bedroom Light"
action_input = "ON"

light = SmartLight(name_input)
light.set_status(action_input)
light.display()
```

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.dept}")

# Execution
emp = Employee("Rahul", "EMP102", "IT")
emp.display_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Execution
c = Contact("Anita", "9876543210")
c.show_contact()
```

```python
class PriceTag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# Execution
p = PriceTag("Headphones", 2499)
p.print_tag()
```

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_rating(self):
        print(f"Movie: {self.title}")
        print(f"Rating: {self.rating} / 5")

# Execution
m = Movie("Inception", 4.8)
m.show_rating()
```

```python
class DeliveryManager:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

# Execution
d = DeliveryManager("Suman", "Hyderabad")
d.print_details()
```
